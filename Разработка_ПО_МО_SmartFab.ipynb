Пример: «авд» это подстрока строки «абвгд», а «адв», «ааб», «вге» - нет.

Алгоритм можно привести на любом языке (предпочтение Python или С/С++) с использованием любых стандартных для этого языка структур данных.

````python
t = 'абвбгде'
s = 'аббде'
def StrInSubstr(t,s):
    sim_cnt = 0
    j = 0
    for i in range(len(s)):
        el_s = s[i]
        #print(el_s)
        while(j<len(t)):
            el_t = t[j]
            j+=1
            if (el_s == el_t):
                sim_cnt+=1
                break
    if sim_cnt == len(s):
        return True
    return False
StrInSubstr(t,s)
````

# Часть 2

Оценить взаимосвязи между признаками с учетом типа признаков.
Написать комментарий почему применяется тот или иной тип рассчета связи между переменными.

Результат: матрица связей (или несколько матриц) приветствуется визуализация в виде тепловых карт

````python
from scipy.stats import f_oneway

def eta_squared(df, group_var, target_var):
    # Удаляем строки с пропусками в обеих переменных
    clean_df = df.dropna(subset=[group_var, target_var])
    
    # Проверяем, что группирующая переменная категориальная
    if clean_df[group_var].nunique() == len(clean_df):
        raise ValueError(f"'{group_var}' должен быть категориальной переменной с повторяющимися значениями")
    
    groups = clean_df.groupby(group_var)[target_var].apply(list)
    
    # Проверка на минимальное количество групп
    if len(groups) < 2:
        return None
    
    F, _ = f_oneway(*groups)
    n_total = len(clean_df)
    n_groups = len(groups)
    
    eta_sq = F * (n_groups - 1) / (F * (n_groups - 1) + (n_total - n_groups))
    return eta_sq
````

````python
import numpy as np
from scipy.stats import chi2_contingency

def cramers_v(x, y):
    """Вычисляет Cramér's V для двух категориальных переменных."""
    confusion_matrix = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min((k-1), (r-1)))
````

````python
v = cramers_v(dataset["Project"], dataset["Grade"])# получается, что 1 из этих признаков не несет никакой информации
print(f"Cramér's V: {v:.3f}")
````
